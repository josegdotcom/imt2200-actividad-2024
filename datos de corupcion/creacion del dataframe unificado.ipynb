### abrimos el archivo que contiene la informacion sobre los casos de corrupcion

```python
df_cc = pd.read_csv("casos_corrupcion_chile.csv", sep=";")

df_cc.info()
print(list(df_cc.columns))
df_cc.head()
```

```python
# primero filtrameromos las comunas para que sean especcificamente de la region metropolitana ademas de quedarme con las columnas necesarias

comunas_santiago = [
    "Cerrillos", "Cerro Navia", "Conchalí", "El Bosque", "Estación Central",
    "Huechuraba", "Independencia", "La Cisterna", "La Florida", "La Granja",
    "La Pintana", "La Reina", "Las Condes", "Lo Barnechea", "Lo Espejo",
    "Lo Prado", "Macul", "Maipú", "Ñuñoa", "Pedro Aguirre Cerda",
    "Peñalolén", "Providencia", "Pudahuel", "Quilicura", "Quinta Normal",
    "Recoleta", "Renca", "San Joaquín", "San Miguel", "San Ramón",
    "Santiago", "Vitacura"
]

df_cc["Comuna"].dropna(axis=0, inplace=True)
df_cc = df_cc[(df_cc["Comuna"].isin(comunas_santiago)) & (df_cc["Año"] >= 2018)].reset_index(drop=True)

df_cc = df_cc[['Año', 'Año inicio', 'Año fin', 'Monto', 'Sector', 'Comuna', 'Posición', 'Perjudicado', 'Delitos']]
df_cc
```
