```python
# importamos la libreria necesaria para la lectura
import pandas as pd

# nos situamos en la carpeta correspondiente
%cd "datos crudos"
```

### Abrimos el archivo que contiene el indice de criminalidad por delito

```python
df_ind_del = pd.read_excel("reportesEstadisticos-porDelito.xlsx", header=22)

df_ind_del.info()
df_ind_del
```

```python
# como mi formato siempre es de que las fechas se encuentren en las filas voy a invertir este dataframe
df_ind_del_inv = df_ind_del.T.reset_index(drop=True).iloc[1:-5]

# le cambiamos el nombre de las columnas de una forma muy original
df_ind_del_inv.columns = ["Delitos violentos", "Delitos asociados a drogas", "Delitos asociados a armas", "Delitos contra la propiedad no violentos"]

fechas = []
mes = 0
ano = 2018
for _ in range(len(list(df_ind_del.columns))):
    mes += 1
    if mes == 13:
        ano += 1
        mes = 0
    fecha = str(ano) + "-" + str(mes).zfill(2)
    fechas.append(fecha)

fechas = pd.DataFrame({"mes": fechas[:-1]})

df_ind_del = pd.merge(fechas, df_ind_del_inv, left_index=True, right_index=True)
df_ind_del
```
