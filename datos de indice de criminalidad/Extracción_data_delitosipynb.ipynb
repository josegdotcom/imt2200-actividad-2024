```python
soup = BeautifulSoup(response.text, "html.parser")
table = soup.find("table")
df = pd.read_html(str(table))[0]
df
```

```python
arrays = [df.values[0], df.values[1], df.values[2],df.values[3],df.values[4],df.values[5]]

deli_data = pd.DataFrame(np.column_stack(arrays))
deli_data.columns = deli_data.iloc[0]
deli_data = deli_data[1:]

fechas = pd.date_range(start='2018-01-01', end='2024-12-01', freq='MS')

deli_data['fecha'] = fechas
deli_data
```
